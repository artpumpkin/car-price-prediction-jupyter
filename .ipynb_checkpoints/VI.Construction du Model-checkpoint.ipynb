<h1 style='color: rgba(0, 0, 0, 0.6); background: whitesmoke; padding: 25px; border-radius: 25px; text-align: center'>VI. Construction du Model</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as matplotlib
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn import decomposition
from sklearn.ensemble import ExtraTreesClassifier, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, KFold
import pickle
```

```python
df = pd.read_csv('MarocAnnoncesDataset_final.csv')
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head(3)
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>1. la visualisation de distribution de prix </h3>

```python
sns.distplot(df.prix)
plt.tight_layout()
plt.title("Distribution du prixe")
plt.savefig('Insight-Dist-Plot',dpi=None)
plt.show()
df.describe()
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>2. correction de la distribution de prix </h3>

```python
#scaling target variable
q1,q3=(df['prix'].quantile([0.25,0.75]))
o1=q1-1.5*(q3-q1)
o2=q3+1.5*(q3-q1)
df2=df[(df.prix>=o1) & (df.prix<=o2)]
```

```python
print(q1," ",q3)
print(o1," ",o2)
print(1.5*(q3-q1))
df2.describe()
```

```python
sns.distplot(df2.prix)
plt.tight_layout()
plt.title("Distribution du prixe")
plt.savefig('Insight-Dist-Plot',dpi=None)
plt.show()
```

```python
df2
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>3. la normalisation des données </h3>

```python
scaler = MinMaxScaler()
scaler.fit(df2)
df2 = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>4. sauvegarder le modele minmax  pour le reutiliser dans l'API REST : </h3>

```python
with open("scaler_minMax.pickle", 'wb') as f:
    pickle.dump(scaler, f)
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>5. diviser le dataset en Y variable target " prix " et les autres variables en X  </h3>

```python
X = df2.drop('prix', axis=1)
Y = df2['prix']
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>6. Diviser le dataset en 80% training 20% pour le test</h3>

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=5)
```


<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>7. Appliquer RFE pour detecter l'importance de chaque variable:</h3> 

```python
from sklearn.feature_selection import RFE

for k in range(2, 17):
    selector = RFE(LinearRegression(), n_features_to_select=k)
    selector.fit(X_train, y_train)
    y_pred = selector.predict(X_test)
    print(f"RFE k = {k} | r2_score = {r2_score(y_test, y_pred)}")
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>8. cette dataframe va contenir des informations de performance pour chaque modéle utilisé </h3>

```python
compareSet=pd.DataFrame(index=['MSE',"RMSE", 'R2 Score','Accuracy(%)',"MSE_Val","RMSE_Val","R2_val","Acc_val"]) 
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>9.Linear Regression :</h3>

```python
LR=LinearRegression()
LR.fit(X_train,y_train)
y_pred=LR.predict(X_test)
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>10.  Feature Importance pour le modele linear régression :  </h3>

```python
coef = pd.Series(LR.coef_, index = X_test.columns)

imp_coef = coef.sort_values()

matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
imp_coef.plot(kind = "barh")

plt.title("Feature importance using Linear Regression Model")
plt.show()
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>11. Calculater les metrics de performance [ 'MSE' , 'RMSE' , 'R2' , 'Accuracy' ] </h3>

```python
MSE=mean_squared_error(y_test, y_pred,squared=True)
RMSE=mean_squared_error(y_test, y_pred,squared=False)
R2=r2_score(y_test,y_pred)
Acc=round(R2*100,2)

print('*'*50)
print(" MSE {} or {}%".format(MSE,round(MSE*100,2)))

print('*'*50)
print(" RMSE {} or {}%".format(RMSE,round(RMSE*100,2)))

print('*'*50)
print(" R2 Score {}".format(R2))

print('*'*50)
print(" Accuracy  {}%".format(Acc))
print('*'*50)
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>12. R2 pour la CrossValidation sur le modele linear regression  : </h3>

```python
from sklearn.model_selection import cross_val_score
import statistics

folds = KFold(n_splits = 5, shuffle = True, random_state = 100)
scores = cross_val_score(LR, X_train, y_train, scoring='r2', cv=folds)
scores
```

```python
R2_val=statistics.mean(scores)
R2_val
```

<h4 style="color:green;">> Visualise the score for each Fold :</h4>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,0.5])
langs = ["Fold N* {}".format(x+1) for x in range(len(scores))]
ax.bar(langs,scores)
plt.title("<< R2 Score for each Fold >>")
plt.show()
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>13. MSE du Cross validation sur le modéle linear regression: </h3> 

```python
Error_val = cross_val_score(LR, X_train, y_train, scoring='neg_mean_squared_error', cv=folds)*(-1)
Error_val
```

```python
MSE_val=statistics.mean(Error_val)
MSE_val
```

<h4 style="color:green;">> Visualise the MSE for each Fold :</h4>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,0.5])
langs = ["Fold N* {}".format(x) for x in range(len(Error_val))]
ax.bar(langs,Error_val)
plt.show()
```

```python
RMSE_val =np.sqrt(MSE_val)
Acc_val=round(R2_val*100,2)


print('*'*60)
print("* MSE {} or {}% in the Validation phase \n*".format(MSE_val,round(MSE_val*100,2)))

print('*'*60)
print("* RMSE in the Validation phase {} or {}% \n*".format(RMSE_val,round(RMSE_val*100,2)))

print('*'*60)
print("* R2 Score {} in the Validation phase \n*".format(R2_val))

print('*'*60)
print("* Accuracy  {}% in the Validation phase \n*".format(Acc_val))
print('*'*60)
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>14. Random Forest Regressor : </h3> 

```python
for k in range(2, 17):
    RFR = RandomForestRegressor(n_estimators=180,random_state=0, min_samples_leaf=1, max_features=0.5, n_jobs=-1, oob_score=True)
    selector = RFE(RFR, n_features_to_select=k)
    selector.fit(X_train, y_train)
    y_pred = selector.predict(X_test)
    print(f"RFE k = {k} | r2_score = {r2_score(y_test, y_pred)}")
```

```python
RFR = RandomForestRegressor(n_estimators=180,random_state=0, min_samples_leaf=1, max_features=0.5, n_jobs=-1, oob_score=True)
RFR.fit(X_train,y_train)
y_pred = RFR.predict(X_test)
```
