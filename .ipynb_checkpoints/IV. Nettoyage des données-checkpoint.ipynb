<h1 style='color: rgba(0, 0, 0, 0.6); background: whitesmoke; padding: 25px; border-radius: 25px; text-align: center'>IV. Nettoyage des données</h1>

```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import KNNImputer
```

```python
df = pd.read_csv('MarocAnnoncesDataset_corrected.csv')
df.drop("Unnamed: 0", axis=1, inplace=True)
# téléphone is converted to float because importing with a csv format can't decide whether the field is a string or a
df.head(3)
```

```python
print("\n".join(df.columns))
```

```python
print(f"DataFrame length: {len(df)}")
```

```python
df.info()
```

```python
def get_nulls():
    nulls = len(df) - df.count()
    nulls = nulls[nulls != 0].sort_values()
    return nulls

get_nulls()
```

```python
nulls_percentage = get_nulls() / len(df) * 100
nulls_percentage
```

```python
plt.figure(figsize=(5, 4))
plt.bar(nulls_percentage.index, nulls_percentage.values)
plt.xticks(rotation=90)
plt.yticks(np.linspace(0, 100, 11), map(lambda x: str(int(x)) + "%", np.linspace(0, 100, 11)))
plt.show()
```

---

```python
def show_missings(attribute):
    number = len(df) - df[attribute].count()
    percentage = round((len(df) - df[attribute].count()) / len(df) * 100, 2)
    print(f"Missing values of {attribute}: {number} | {percentage}%")
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>3. marque</h3>

```python
show_missings("marque")
```

```python
df["marque"].value_counts()
```

```python
sf = df[df["marque"].isnull()][["nom d'annonce", "description", "marque", "prix"]]
sf
```

```python
sf[sf["prix"] < 20000]
```

```python
df = df[df["prix"] > 20000]
sf = df[df["marque"].isnull()][["nom d'annonce", "description", "marque", "prix"]]
sf
```

```python
sf = sf[sf["nom d'annonce"].str.contains(r"bate|camion|camm|jet|tracteur|moteur") | sf["description"].str.contains(r"bate|camion|camm|jet|tracteur|moteur")]
sf
```

```python
sf.index
```

```python
df
```

```python
df.drop(sf.index, inplace=True)
```

```python
marques = df[df["marque"].notnull()]["marque"].unique()
marques
```

```python
def clean_marque(row):
    if row["marque"] == row["marque"]:
        return row
    
    match = re.search("|".join(marques), row["nom d'annonce"] + row["description"])
    if match:
        row["marque"] = match.group()
        return row
    return row

df = df.apply(clean_marque, axis=1)
df.head(3)
```

```python
show_missings("marque")
```

```python
df["marque"] = df["marque"].apply(lambda x: 'NaN' if x != x else x)
df["marque"]
```

```python
encoder = LabelEncoder()
output = encoder.fit_transform(df["marque"]) 
output
```

```python
encoder.classes_
```

```python
df["marque"] = output
df["marque"] = df["marque"].apply(lambda x: None if x == 0 else x)
df["marque"] 
```

```python
df.head(4)
```

```python
def use_KNNImputer(attribute, n_neighbors):
    imputer = KNNImputer(n_neighbors=n_neighbors)
    result = imputer.fit_transform(df[attribute].to_numpy().reshape(-1, 1)).round()
    df[attribute] = result
```

```python
use_KNNImputer('marque', 5)
```

```python
show_missings("marque")
```

```python
df["marque"].unique()
```

```python
df["marque"] = df["marque"].apply(lambda x: encoder.classes_[int(x)])
df["marque"]
```

```python
dummies = pd.get_dummies(df["marque"])
dummies.drop("tvr", axis=1, inplace=True)
dummies.head(5)
```

```python
df = pd.concat([df, dummies], axis=1).reset_index()
df.head(5)
```

```python
df.drop(["index", "marque"], axis=1, inplace=True)
df.head(5)
```
