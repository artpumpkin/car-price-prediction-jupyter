<h1 style='color: rgba(0, 0, 0, 0.6); background: whitesmoke; padding: 25px; border-radius: 25px; text-align: center'>V. Ingénierie des caractéristiques</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import decomposition
from sklearn.ensemble import ExtraTreesClassifier
```

```python
df = pd.read_csv('MarocAnnoncesDataset_cleaned.csv')
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head(3)
```

```python
print(f"DataFrame length: {len(df)}")
```

```python
df.info()
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>1. Feature Importance</h3>

```python
classifier = ExtraTreesClassifier()
classifier.fit(df.drop("prix", axis=1), df["prix"])
```

```python
extratrees = pd.DataFrame(classifier.feature_importances_, columns=["extratrees"])
columns = pd.DataFrame(df.drop("prix", axis=1).columns, columns=["columns"])

result = pd.concat([extratrees, columns], axis=1)
result
```

```python
pd.set_option('display.max_rows', 100)
result.sort_values(['extratrees'], ascending=False)
```

```python
limited_results = result[result["extratrees"] >= 0.001]
limited_results
```

```python
df.drop(result[result["extratrees"] < 0.001]["columns"], axis=1, inplace=True)
df.head()
```

<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>2. Réduction de dimensionnalité</h3>

```python
X = df.drop("prix", axis=1)
X.head()
```

```python
scaler = StandardScaler()
normalized = scaler.fit_transform(X)

X = pd.DataFrame(normalized, columns=X.columns)
X.head()
```

```python
pca = decomposition.PCA(n_components=2)

pca.fit(X)

# explained_variance_ : array, shape (n_components,)
#     The amount of variance explained by each of the selected components.

#     Equal to n_components largest eigenvalues
#     of the covariance matrix of X.

#     .. versionadded:: 0.18

# explained_variance_ratio_ : array, shape (n_components,)
#     Percentage of variance explained by each of the selected components.

#     If ``n_components`` is not set then all components are stored and the
#     sum of the ratios is equal to 1.0.

# print(pca.components_)
print(f"explained variance              : {pca.explained_variance_}")
print(f"explained variance ratio        : {pca.explained_variance_ratio_}")
print(f"explained variance ratio cumsum : {pca.explained_variance_ratio_.cumsum()}")
print(f"number of components            : {pca.n_components_}")
print(f"number of features              : {pca.n_features_}")
print(f"number of samples               : {pca.n_samples_}")
```

```python
projected_X = pca.transform(X)
projected_X
```

```python
projected_X.shape
```

```python
plt.figure(figsize=(10,7))
plt.scatter(projected_X[:, 0], projected_X[:, 1], c=df['prix'], alpha=0.5)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
pca.components_
```

```python
pca.components_.shape
```

```python
pc = pca.components_

plt.figure(figsize=(10,10))
for i, (x, y) in enumerate(zip(pc[0, :], pc[1, :])):
    plt.plot([0, x], [0, y])
    plt.text(x, y, X.columns[i])
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
def already_in(document, characteristic):
    for key in document:
        for name in document[key]:
            if (characteristic == name):
                return True
    return False

def get_columns_2_keep(pc, df, threshold):
    document = dict()
    for i, (x, y) in enumerate(zip(pc[0, :], pc[1, :])):
        if already_in(document, df.columns[i]):
            continue

        document[df.columns[i]] = []
        for j, (x2, y2) in enumerate(zip(pc[0, :], pc[1, :])):
            if df.columns[i] != df.columns[j]:
                if (x - x2) ** 2 + (y - y2) ** 2 < threshold:
                    document[df.columns[i]].append(df.columns[j])
    return document.keys()
```

```python
columns_2_keep = get_columns_2_keep(pc, X, 0.01)

columns_2_keep
```

```python
df.drop([column for column in X.columns if column not in columns_2_keep], axis=1, inplace=True)
df.head()
```
