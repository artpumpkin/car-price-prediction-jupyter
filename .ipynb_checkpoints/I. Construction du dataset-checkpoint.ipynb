<h3 style='color: rgba(0, 0, 0, 0.6); background: #fddddb; padding: 10px; border-radius: 25px;'>1.  Réalisation du scrapping</h3>

```python
# https://www.marocannonces.com/categorie/15/Auto-Moto/[min=1,max=2242].html
# https://www.marocannonces.com/index.php?a=14&b=[id]
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
links = []
MAX = 2243
i = 1
while i < MAX:
    print(i)
    try:
        document = requests.get(f"https://www.marocannonces.com/categorie/15/Auto-Moto/{i}.html")
        soup = BeautifulSoup(document.content, 'html.parser')
        spans = soup.find_all('span', title='Favoris')
        for span in spans:
            id = span.get('rel')
            link = f"https://www.marocannonces.com/index.php?a=14&b={id}"
            links.append(link)        
        i += 1
    except Exception as e:
        print(e)
print('done')
```

```python
links
```

```python
print(f"There is more than {len(links)} link!")
```

```python
features = ["id", "nom d'annonce", "nom d'annonceur", "téléphone", "prix", "date d'ajout", "ville", "visite", "année", "kilométrage", "carburant", "puissance", "marque", "modèle", "description"]
features
```

```python
documents = dict()
for feature in features:
    documents[feature] = []
documents
```

```python
i = 44000
# MAX = len(links) # 44830 link
MAX = 44830
while i < MAX:
    print(i)
    try:
        link = links[i]
        document = requests.get(link)
        soup = BeautifulSoup(document.content, 'html.parser')
        
        # # id
        documents["id"].append(link[len(link) - 7:])
        
        # # nom d'annonce
        try:
            result = soup.select('.listing_title')[0].text.split("\r\n\t\t\t")[1].replace('\r\n\t\t', '').strip()
            documents["nom d'annonce"].append(result)
        except Exception as e:
            documents["nom d'annonce"].append(None)
        
        # nom d'annonceur
        try:
            result = re.sub('^\s*', '', soup.select('.seller_username')[0].text.strip().split("\n")[1])
            documents["nom d'annonceur"].append(result)
        except Exception as e:
            documents["nom d'annonceur"].append(None)

        # téléphone
        try:
            result = soup.select('.vertical_list')[0].text.strip().replace('Téléphone : ', '')
            documents["téléphone"].append(result)
        except Exception as e:
            documents["téléphone"].append(None)

        # prix
        try:
            result = soup.select('.value.price')[0].text
            documents["prix"].append(result)
        except Exception as e:
            documents["prix"].append(None)

        # date d'ajout
        try:
            result = soup.select('#print_listing_info_column .info > li.value:nth-of-type(4)')[0].text
            documents["date d'ajout"].append(result)
        except Exception as e:
            documents["date d'ajout"].append(None)

        # ville
        try:
            result = soup.select('#print_listing_info_column .info > li.value:nth-of-type(8)')[0].text
            documents["ville"].append(result)
        except Exception as e:
            documents["ville"].append(None)

        # visite
        try:
            result = soup.select('.row_even')[0].text.replace('Visites: ', '').strip()
            documents["visite"].append(result)
        except Exception as e:
            documents["visite"].append(None)


        annee = None
        kilometrage = None
        carburant = None
        puissance = None
       
        try:
            for li in soup.select('#extra_questions li'):
                # année
                if 'Année' in li.text:
                    annee = li.text.replace('Année : ', '')
                
                # kilométrage
                elif 'Kilométrage' in li.text:
                    kilometrage = li.text.replace('Kilométrage : ', '')
                
                # carburant
                elif 'Carburant' in li.text:
                    carburant = li.text.replace('Carburant  : ', '')
                
                # puissance
                elif 'Puissance' in li.text:
                    puissance = li.text.replace('Puissance : ', '')

        except Exception as e:
            pass

        documents["année"].append(annee)
        documents["kilométrage"].append(kilometrage)
        documents["carburant"].append(carburant)
        documents["puissance"].append(puissance)


        # marque
        try:
            result = soup.select('.print_half_column_left .info li:nth-of-type(1)')[0].text.replace('Marque: ', '')
            documents["marque"].append(result)
        except Exception as e:
            documents["marque"].append(None)

        # modèle
        try:
            result = soup.select('.print_half_column_left .info li:nth-of-type(2)')[0].text.replace('Modèle: ', '')
            documents["modèle"].append(result)
        except Exception as e:
            documents["modèle"].append(None)

        # description
        try:
            result = soup.select('.box_pad')[0].text.strip()
            documents["description"].append(result)
        except Exception as e:
            documents["description"].append(None)
            
        i += 1
    except Exception as e:
        print(e)
    
print('done')
```

### id

> This attribute signifies the name of the car post announcer.

> This attribute signifies the price of the car post.

### date d'ajout

### ville

> This attribute signifies the city of the car post.

### visite

> This attribute signifies the number of visitors of the car post.

### année

> This attribute signifies the production date of the car.

###  kilométrage

> This attribute signifies the number of kilometrage traveled or covered by the car.

### carburant

> This attribute signifies the type of fuel used by the car.

### puissance
> This attribute signifies the fiscal power of the car.

### marque

> This attribute signifies the marque of the car post.

### modèle

> This attribute signifies the brand of the car post.

###  description
> This attribute signifies the description of the car post.
