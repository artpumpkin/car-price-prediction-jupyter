```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MarocAnnoncesDataset.csv')
df.head(3)
```

```python
df.info()
```

```python
nulls = len(df) - df.count()
nulls = nulls[nulls != 0].sort_values()
nulls
```

```python
plt.figure(figsize=(8, 4))
plt.bar(nulls.index, nulls.values)
plt.xticks(rotation=90)
plt.show()
```

In this graph we can observe that the **kilométrage** has the most missing values of 17500 and **nom d'annonce** has the least of them of 16.

We can also notice that our target **prix** has a number of missing values of 6421 which will eventually delete.
